# how to merge

```python
a = pd.DataFrame({'cat_name':['elctro','kitchien','charir'],'cat_id':[0,1,2]})
```

```python
b = pd.DataFrame({'name':['mac','painting','cup','chir'],'cat_id':[0,5,2],'item_id':[0,1,2]})
```

## the merge ( inner , outer , right , left )


```python
i = pd.merge(a , b , how='inner' , on = 'cat_id' )
i
```

```python
o = pd.merge(a , b , how= 'outer' , on= 'cat_id')
o
```

```python
a
```

```python
b
```

```python
r = pd.merge(a,b,how='right' , on= 'cat_id')
r
```

```python
l = pd.merge(a,b,how='left' , on= 'cat_id')
l
```

```python
x = pd.DataFrame({'cat_name':['elctro','kitchien','coffe'],'cat_id':[0,1,2]})
x
```

```python
z = pd.DataFrame({'name':['mac','painting','cup','chair'],'cat_id':[0,5,2,4],'item_id':[0,1,2,3]})
z
```

```python
right =   pd.merge(z,x,how='right' , on= 'cat_id')
right
```

```python
left  =  pd.merge(x,z,how='left' , on= 'cat_id')
left
```

```python
left = pd.DataFrame({'key':['foo','bar'],'val':[1,2]})
left
```

```python
right = pd.DataFrame({'key':['foo','bar'],'val':[4,5]})
right
```

```python
pd.merge(left , right , how="inner" , on='key') #   _y  _x
```

```python
right = pd.DataFrame({'key_right':['foo','bar'],'val':[4,5]})
right
```

```python
left = pd.DataFrame({'key_left':['foo','bar'],'val':[1,2]})
left
```

```python
pd.merge(left , right , how="inner" ,suffixes=('_from_left','_from_right') , left_on='key_left' , right_on='key_right' ).drop(columns='key_right')
```
