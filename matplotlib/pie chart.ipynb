```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
#plt.style.use('')
```

```python
data = [70,20,10,50]
labels = ['Python', 'Javascript', 'HTML', 'Java']
explode = [0.1,0,0.2,0]
plt.pie(data,labels=labels,wedgeprops={'edgecolor':'black'})
plt.title('pie chart')
plt.show()
```

```python
data = [70,20,10,50]
labels = ['Python', 'Javascript', 'HTML', 'Java']
explode = [0.1,0,0.2,0.3]
plt.pie(data,labels=labels,explode=explode,wedgeprops={'edgecolor':'black'},shadow=True,startangle=90,autopct='%1.1f%%')
plt.title('pie chart')
plt.show()
```
