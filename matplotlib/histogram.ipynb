```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-v0_8')
```

```python
plt.style.available
```

```python
ages = [10, 30, 45, 30, 31, 46, 12, 16, 32, 23, 28, 29]
bins = [10,20,30,40,50,60]
mean = np.mean(ages)

plt.hist(ages,bins=bins,edgecolor='black',linewidth= 3) 
plt.axvline(mean,color='r',linewidth= 6)

plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
